# Data Loading

```python
import pandas as pd
import os

data_dir = "../data"
#Read the data from data folder
user_sample = pd.read_json(os.path.join(data_dir, "user_sample.json"), lines=True)
edge_sample = pd.read_csv(os.path.join(data_dir, "edge_sample.csv"))
label_sample = pd.read_csv(os.path.join(data_dir, "label_sample.csv"))
```

# Social Network Analysis based on Following Graph

```python
import networkx as nx
G = nx.Graph()  # Create an undirected graph


for index, row in edge_sample.iterrows():
    if row['relation'] == 'following':
        G.add_edge(row['source_id'], row['target_id']) # Add following relationships into the graph
    
```

```python
label = {}
for index, row in label_sample.iterrows():
    label[row['id']] = row['label'] # Transform the label of users into dict
```

```python
# Compute centrality measures and structure hole measures
degree_centrality = nx.degree_centrality(G)  # Degree Centrality
betweenness_centrality = nx.betweenness_centrality(G)  # Betweenness Centrality
closeness_centrality = nx.closeness_centrality(G)  # Closeness Centrality
effective_size = nx.effective_size(G) # Effective Size
```

```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import os

# Ensure visualizations directory exists
visualizations_dir = "../visualizations"
os.makedirs(visualizations_dir, exist_ok=True)

# Function to separate values by label
def get_centrality_by_label(centrality_dict, label_dict):
    human_values = []
    bot_values = []
    for node, value in centrality_dict.items():
        if node in label_dict:
            if label_dict[node] == 'human':
                human_values.append(value)
            else:
                bot_values.append(value)
    return np.array(human_values), np.array(bot_values)

# Compute centrality measures
degree_centrality = nx.degree_centrality(G)  # Degree Centrality
betweenness_centrality = nx.betweenness_centrality(G)  # Betweenness Centrality
closeness_centrality = nx.closeness_centrality(G)  # Closeness Centrality
effective_size = nx.effective_size(G)  # Effective Size

# Get centrality values for humans and bots
deg_human, deg_bot = get_centrality_by_label(degree_centrality, label)
bet_human, bet_bot = get_centrality_by_label(betweenness_centrality, label)
clo_human, clo_bot = get_centrality_by_label(closeness_centrality, label)
eff_human, eff_bot = get_centrality_by_label(effective_size, label)

# Function to plot CDF and save it
def plot_cdf(data1, data2, title, xlabel, filename):
    plt.figure(figsize=(6, 4))
    
    # Sort data for CDF
    x1 = np.sort(data1)
    y1 = np.arange(1, len(x1) + 1) / len(x1)
    
    x2 = np.sort(data2)
    y2 = np.arange(1, len(x2) + 1) / len(x2)
    
    plt.plot(x1, y1, label="Human", linewidth=2)
    plt.plot(x2, y2, label="Bot", linewidth=2)
    
    plt.xlabel(xlabel)
    plt.ylabel("CDF")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    
    save_path = os.path.join(visualizations_dir, 'Explanation_' + filename)
    plt.savefig(save_path)
    plt.show()
    plt.close()

# Plot and save CDFs
plot_cdf(deg_human, deg_bot, "Degree Centrality CDF", "Degree Centrality", "degree_centrality_cdf.png")
plot_cdf(bet_human, bet_bot, "Betweenness Centrality CDF", "Betweenness Centrality", "betweenness_centrality_cdf.png")
plot_cdf(clo_human, clo_bot, "Closeness Centrality CDF", "Closeness Centrality", "closeness_centrality_cdf.png")
plot_cdf(eff_human, eff_bot, "Effective Size CDF", "Effective Size", "effective_size_cdf.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import os

# Ensure visualizations directory exists
visualizations_dir = "../visualizations"
os.makedirs(visualizations_dir, exist_ok=True)

# Function to separate values by label
def get_centrality_by_label(centrality_dict, label_dict):
    human_values = []
    bot_values = []
    for node, value in centrality_dict.items():
        if node in label_dict:
            if label_dict[node] == 'human':
                human_values.append(value)
            else:
                bot_values.append(value)
    return np.array(human_values), np.array(bot_values)

# Get centrality values for humans and bots
deg_human, deg_bot = get_centrality_by_label(degree_centrality, label)
bet_human, bet_bot = get_centrality_by_label(betweenness_centrality, label)
clo_human, clo_bot = get_centrality_by_label(closeness_centrality, label)
eff_human, eff_bot = get_centrality_by_label(effective_size, label)

# Function to plot CDF in a subplot grid
def plot_cdf(ax, data1, data2, title, xlabel):
    # Sort data for CDF
    x1 = np.sort(data1)
    y1 = np.arange(1, len(x1) + 1) / len(x1)
    
    x2 = np.sort(data2)
    y2 = np.arange(1, len(x2) + 1) / len(x2)
    
    ax.plot(x1, y1, label="Human", linewidth=2)
    ax.plot(x2, y2, label="Bot", linewidth=2)
    
    ax.set_xlabel(xlabel)
    ax.set_ylabel("CDF")
    ax.set_title(title)
    ax.legend()
    ax.grid(True)

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot CDFs in the subplots
plot_cdf(axs[0, 0], deg_human, deg_bot, "Degree Centrality CDF", "Degree Centrality")
plot_cdf(axs[0, 1], bet_human, bet_bot, "Betweenness Centrality CDF", "Betweenness Centrality")
plot_cdf(axs[1, 0], clo_human, clo_bot, "Closeness Centrality CDF", "Closeness Centrality")
plot_cdf(axs[1, 1], eff_human, eff_bot, "Effective Size CDF", "Effective Size")

# Adjust layout
plt.tight_layout()

# Save the combined plot
save_path = os.path.join(visualizations_dir, "Explanation_network_metrics_cdfs.png")
plt.savefig(save_path)

# Show the combined plot
plt.show()

plt.close()
```

```python
import seaborn as sns
import os
import matplotlib.pyplot as plt

# Ensure visualizations directory exists
visualizations_dir = "../visualizations"
os.makedirs(visualizations_dir, exist_ok=True)

# Function to plot distribution (Histogram + KDE) and save it
def plot_distribution(data1, data2, title, xlabel, filename):
    plt.figure(figsize=(6, 4))

    # Histogram with KDE
    sns.histplot(data1, kde=True, color='blue', label='Human', bins=30, stat='density', alpha=0.6)
    sns.histplot(data2, kde=True, color='red', label='Bot', bins=30, stat='density', alpha=0.6)
    
    plt.xlabel(xlabel)
    plt.ylabel("Density")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    
    # Save the plot
    save_path = os.path.join(visualizations_dir, 'Explanation_' + filename)
    plt.savefig(save_path)
    
    # Show the plot
    plt.show()
    
    plt.close()

# Plot and save distributions
plot_distribution(deg_human, deg_bot, "Degree Centrality Distribution", "Degree Centrality", "degree_centrality_dist.png")
plot_distribution(bet_human, bet_bot, "Betweenness Centrality Distribution", "Betweenness Centrality", "betweenness_centrality_dist.png")
plot_distribution(clo_human, clo_bot, "Closeness Centrality Distribution", "Closeness Centrality", "closeness_centrality_dist.png")
plot_distribution(eff_human, eff_bot, "Effective Size Distribution", "Effective Size", "effective_size_dist.png")
```

# **Results Analysis**

The following CDF plots compare the distributions of **degree centrality, betweenness centrality, closeness centrality, and effective size** between humans and bots in the Twibot-22 dataset.

## **1. Degree Centrality**  
- Both humans and bots have low degree centrality, with most values near **zero**.  
- Bots (orange) accumulate faster, indicating **fewer connections** compared to humans.  
- This suggests that bots are **less integrated** into the network.  

## **2. Betweenness Centrality**  
- Highly skewed distribution, with most accounts having low values.  
- Bots tend to have **lower betweenness**, meaning they are **less likely to be key intermediaries** in the network.  
- This implies that bots do not act as central hubs in information flow.  

## **3. Closeness Centrality**  
- Shows a **more gradual distribution** compared to the other measures.  
- Bots accumulate faster at **lower values**, indicating they are **farther from other nodes** and more **isolated**.  

## **4. Effective Size**  
- Humans generally have a **larger effective size** than bots.  
- This suggests that humans occupy more **structural holes**, meaning they have **more diverse and non-redundant connections** in the network.  
- Bots, on the other hand, tend to form **redundant or clustered connections**, reducing their effective network reach.  

## **Key Insights**  
- Bots generally have **lower degree, betweenness, and closeness centrality**, making them **less connected and less influential**.  
- The distinction is clearer in **degree and betweenness centrality**, while closeness centrality shows a **more gradual difference**.  
- **Effective size provides additional insight**, showing that humans are more likely to bridge different parts of the network, while bots form **more redundant connections**.  
- These findings suggest that **network metrics**, particularly effective size, could be useful for bot detection**, especially when combined with other features.
